```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
```

```python
data = pd.read_csv('train.csv')
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
nombres_conversations = len(data)
nombres_patients = data['Context'].nunique()
nombres_therapeutes = data['Response'].nunique()
print(f'Nombre de conversations : {nombres_conversations}')
print(f'Nombre de patients : {nombres_patients}')
print(f'Nombre de therapeutes : {nombres_therapeutes}')
```

```python
data["Context"] = data["Context"].str.lower()
data["Response"] = data["Response"].str.lower()
```

```python
data["Context"] = data["Context"].str.replace('[^\w\s]', '',regex=True)
data["Response"] = data["Response"].str.replace('[^\w\s]', '',regex=True)
```

```python
data["Context"] = data["Context"].str.replace('\d', '')
data["Response"] = data["Response"].str.replace('\d', '')
```

```python
sw = stopwords.words('english')
```

```python
def remove_stopwords(text):
    words = str(text).split()
    filtered_words = []
    for word in words:
        if word not in sw:
            filtered_words.append(word)
    return " ".join(filtered_words)
```

```python
data["Context"] = data["Context"].apply(remove_stopwords)
data["Response"] = data["Response"].apply(remove_stopwords)
```

```python
import re
def tokenization(text):
    return re.findall(r'\b\w+\b', text)
```

```python
data['Context_tokens'] = data['Context'].astype(str).apply(tokenization)
data['Response_tokens'] = data['Response'].astype(str).apply(tokenization)
```

```python
# Statistiques des mots
contextDic = []
responseDic = []
allDic = []
for word in data['Context_tokens']:
    contextDic.extend(word)
    allDic.extend(word)
for word in data['Response_tokens']:
    responseDic.extend(word)
    allDic.extend(word)
from collections import Counter
contextCounter = Counter(contextDic)
responseCounter = Counter(responseDic)
allCounter = Counter(allDic)
```

```python
motsContext = contextCounter.most_common(100)
mots = []
frequences = []
for mot, freq in motsContext:
    mots.append(mot)
    frequences.append(freq)
for mot, freq in motsContext:
    print(f'{mot}: {freq}')
```

```python
motsResponse = responseCounter.most_common(100)
mots = []
frequences = []
for mot, freq in motsResponse:
    mots.append(mot)
    frequences.append(freq)
for mot, freq in motsResponse:
    print(f'{mot}: {freq}')
```

```python
allWords = allCounter.most_common(100)
mots = []
frequences = []
for mot, freq in allWords:
    mots.append(mot)
    frequences.append(freq)
for mot, freq in allWords:
    print(f'{mot}: {freq}')
```

```python
# Fréquence des mots dans chaque conversion pour les Contextes
data['Context_word_count'] = data['Context_tokens'].apply(len)
# Fréquence des mots dans chaque conversion pour les Réponses
data['Response_word_count'] = data['Response_tokens'].apply(len)
# Affichage des statistiques
print(data[['Context_word_count', 'Response_word_count']].describe())
```

```python
data["Context"] = data["Context"].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
data["Response"] = data["Response"].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
```

```python
#Déterminer les sujets méthode 1 : 
stress = ['stress','pressure','tension','fatigue','restlessness','anxiety']
depression = ['depression','sadness','guilt','suicide','insomnia']
```

```python
contexts = data.drop_duplicates(subset=["Context"])
def topic_detection(text,topic):
    n=0
    for sentence in text:
        words = sentence.split()
        count = 0
        for w in words:
            if w in topic:
                count+=1
        if count>0:
            n+=1
    return n
```

```python
stress_count = topic_detection(contexts["Context"], stress)
depression_count = topic_detection(contexts["Context"], depression)
print(f"Stress count: {stress_count}")
print(f"Depression count: {depression_count}")
```

```python
stress_count = topic_detection(contexts["Response"], stress)
depression_count = topic_detection(contexts["Response"], depression)
print(f"Stress count: {stress_count}")
print(f"Depression count: {depression_count}")
```

```python
# Déterminer les sujets méthode 2 :
from sklearn.feature_extraction.text import TfidfVectorizer
df = data["Context"]
vectorizer = TfidfVectorizer(max_features=1000,stop_words='english')
X = vectorizer.fit_transform(df)
feature_names = vectorizer.get_feature_names_out()
scores = np.mean(X.toarray(), axis=0)
top_words = sorted(zip(feature_names, scores), key=lambda x: x[1], reverse=True)
for word,score in top_words[:20]:
    print(word,score)
#Les sujets les plus fréquents sont : 
# Anxiété et dépression : im, help, anxiety
# Problèmes relationnels : relationship, year, love, friend 
# Perte de repères : feel, like, know
```

```python
df = data["Response"]
vectorizer = TfidfVectorizer(max_features=1000,stop_words='english')
X = vectorizer.fit_transform(df)
feature_names = vectorizer.get_feature_names_out()
scores = np.mean(X.toarray(), axis=0)
top_words = sorted(zip(feature_names, scores), key=lambda x: x[1], reverse=True)
for word,score in top_words[:20]:
    print(word,score)
#Les sujets les plus fréquents sont :
# Relations et émotions (Compendre les émotions, les relations avec les autres) : like, feel, know, relationship, year, love
# Développement personnel : therapist, help, workn need
# Décisions et avenir : time, life, want, know, good
```

```python
# Déterminer les sujets méthode 3 :
from sentence_transformers import SentenceTransformer # type: ignore
from sklearn.cluster import KMeans
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(data['Context'].tolist())
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
labels = kmeans.fit_predict(embeddings)
data['Topic_Context'] = labels
for i in range(num_clusters):
    print(f"Sujet {i}:")
    texts = data[data['Topic'] == i]['Context']
    vectorizer = TfidfVectorizer(max_features=10, stop_words='english')
    X = vectorizer.fit_transform(texts)
    print(vectorizer.get_feature_names_out())
# Sujet 0 : Relations personnelles (Solitude)
# Sujet 1 : Problèmes de santé mentale (Anxiété et dépression)
# Sujet 2 : Demande de l'aide (Recherche de soutien et de conseils)
# Sujet 3 : Problèmes relationnels (Difficultés dans les relations amoureuses et amicales)
# Sujet 4 : Problèmes de passé (Traumatismes et expériences passées)
```

```python
embeddings = model.encode(data['Response'].tolist())
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
labels = kmeans.fit_predict(embeddings)
data['Topic_Response'] = labels
for i in range(num_clusters):
    print(f"Sujet {i}:")
    texts = data[data['Topic_Response'] == i]['Response']
    vectorizer = TfidfVectorizer(max_features=10, stop_words='english')
    X = vectorizer.fit_transform(texts)
    print(vectorizer.get_feature_names_out())
# Sujet 0 : Gestion des émotions
# Sujet 1 : Relations et bien-être
# Sujet 2 : Conseils sur les relations
# Sujet 3 : Accompagnement thérapeutique
# Sujet 4 :  Relations familiales
```

```python
# Déterminer les sujets méthode 4 :
from sklearn.decomposition import LatentDirichletAllocation
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X = vectorizer.fit_transform(data['Context'])
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(X)
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lda.components_):
    print(f"Sujet {topic_idx}:")
    top_features_indices = topic.argsort()[:-11:-1]
    top_features = [feature_names[i] for i in top_features_indices]
    print(top_features)
```
